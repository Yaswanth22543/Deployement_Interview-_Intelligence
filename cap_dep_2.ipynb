```python
# Splitting into numcols and objcols
numcols=combinedf.select_dtypes(include=np.number)
objcols=combinedf.select_dtypes(include='object')
```

```python
objcols = objcols.drop('Status',axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

mm=MinMaxScaler()

numcols_mm=mm.fit_transform(numcols)

numcols_mm=pd.DataFrame(numcols_mm,columns=numcols.columns)

numcols_mm.head()
```

```python
objcols.head()
```

```python
objcols.Profile.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Define the columns to be label encoded
columns_to_encode = ['Profile', 'P.E.I', 'P.E.C', 'Interviewer Intro', 'Candidate into', 'Opp to ask']

# Apply Label Encoding to each column
for col in columns_to_encode:
    objcols[col] = label_encoder.fit_transform(objcols[col])

# Now, 'data' will have new columns with '_encoded' suffix containing numerical labels
```
